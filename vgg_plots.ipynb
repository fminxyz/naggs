```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "./vgg11_result_25.04.2023_06:45:52.npz"
filename_adam = "./vgg11_result_25.04.2023_09:13:53_adam.npz"
data = np.load(filename, allow_pickle=True)
data2 = np.load(filename_adam, allow_pickle=True)
```

```python
list(data.keys())
```

```python
for lr in data["test_acc_lr"].item():
    print(lr)
    for m in data["test_acc_lr"].item()[lr]:
        print(m, data["test_acc_lr"].item()[lr][m][-1])
```

```python
for lr in data["lr_range"]:
    plt.figure()
    for method in data["test_acc_lr"].item()[lr]:
        cur_data = [val.cpu() for val in data["test_acc_lr"].item()[lr][method]]
        plt.plot(cur_data, label=method)
    for method in data2["test_acc_lr"].item()[lr]:
        cur_data = [val.cpu() for val in data2["test_acc_lr"].item()[lr][method]]
        plt.plot(cur_data, label=method)
    plt.title("lr = {}".format(lr))
    plt.legend()
```
